### The model

```python
def linear_model(X, beta):
    import numpy as np
    return np.dot(X, beta)
```

### The estimation

```python
def ols(X):
    """
    The matrix which solves the OLS regression problem for full-rank X
    """
    return np.dot(la.pinv(np.dot(X.T, X)), X.T)
```

```python
# We don't have beta or Y for now, so let us fake values for beta
beta = [random.uniform(0.3,0.7) for i in range(0,14)]
Y = linear_model(X, beta)
Y, beta
```

```python
noise_level = 0.01
beta_hat = np.dot(ols(X + np.random.randn(N,p)*noise_level), Y)

def estimate_beta(X, Y, noise_level=0):
    beta_hat = np.dot(ols(X + np.random.randn(N,p)*noise_level), Y)
    return beta_hat
```

```python
plot(beta, beta_hat, 'or')
xlabel(r'$\beta$', fontsize=20)
ylabel(r'$\hat{\beta}$', fontsize=20)
m = min([min(beta), min(beta_hat)])
M = max([max(beta), max(beta_hat)])
plot([m,M], [m,M], '-')
```

```python
errors = []
for i in range(0,10000):
    beta_hat = estimate_beta(X, Y, noise_level=0.01)
    errors.append(beta_hat)
```

### Effect on Y of noisy X

```python
def predict(X, beta, noise_level=0):
    this = linear_model(X+np.random.randn(N,p)*noise_level, beta)
    return this
```

```python
yerrors = []
for i in range(0,10000):
    y = predict(X, beta, noise_level=0.2) #0.1 leads to CV of 0.06
    yerrors.append(y)
yerrors = pd.DataFrame(yerrors)    
```
