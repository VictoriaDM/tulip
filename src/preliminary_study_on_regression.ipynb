### Here, we compute X given Y and beta

```python
def getX(Y, beta):
    beta = np.matrix(beta)
    betaside =  np.dot(la.pinv(np.dot(beta.T, beta)), beta.T)
    return np.dot(betaside, np.matrix(Y))
```

```python
X2 = getX(Y2, beta2)
X2 = X2.T
imshow(X2, aspect='auto', cmap='jet');
```

```python
# Now, we perturb X and compute Y. The idea is to estimate the propagation of the errors from X to Y if
# we know beta very well
yerrors = []
for i in range(0,10000):
    y = predict(X2, beta2, noise_level=0.1) #0.1 leads to CV of 0.06
    yerrors.append(np.array(y)[0])
    
yerrors = pd.DataFrame(yerrors)    
```
