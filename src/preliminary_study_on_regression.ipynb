### Here, we compute X given Y and beta

```python
Y2 = Y.reshape((N,1))
beta2 = beta.reshape((p,1))
##Y2 = array([1,3,5]).reshape((3,1))
#beta2 = array([1,2,1]).reshape((3,1))
```

```python
def getX(Y, beta):
    betaprime = la.pinv(np.dot(beta, beta.T))
    betaprime = np.dot(beta.T, betaprime)
    # Y should be transformed here but numpy artifact ?
    X = np.dot(Y, betaprime)
    return X
```

```python
X2 = getX(Y2, beta2[0:])
# Here we need to force the first column to be one, and get rid of the last one
X2 = np.insert(X2, 0, [1]*14, axis=1)
X2 = X2[:,0:-1]
X2
# Now, we construct a new beta set from Y2 and X2. A bit convoluated but now we have X first first column as ones,
# rest of the columns fully correlated i.e. no noise
beta3 = linreg.estimate_beta(X2, Y2)
Y2
```

```python
# Now, we perturb X and compute Y. The idea is to estimate the propagation of the errors from X to Y if
# we know beta very well
yerrors = []
for i in range(0,100):
    y = predict(X2, beta3, noise_level=1) #0.1 leads to CV of 0.06
    yerrors.append(y.reshape(N,))
    
yerrors = pd.DataFrame(yerrors)    
```

```python
#i=13; hist(yerrors[i],100); axvline(Y[i], lw=2, color='r')
_ = yerrors.hist(bins=100)
axes = gcf().get_axes()
for i in range(0, len(Y2)):
    axes[i].axvline(Y2[i], color='r')
```

```python
def get_score(y):
    # tricky to handle vector vs columns so we convert to make sure we have the correct orientation
    correct = np.array(list(Y2)).T
    r1 = ((correct-y)**2).sum()
    r2 = ((y-correct.mean() )**2).sum()
    return 1-r1/r2
```

```python
scores = [get_score(this[1].values) for this in yerrors.iterrows()]
hist(scores)
mean(scores), std(scores)
```
